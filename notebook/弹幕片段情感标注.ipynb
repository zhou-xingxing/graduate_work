```python
# é¿å…åˆ·å±
day1=pd.read_csv(r'../data/cleaned_room36252danmu0318.csv',header=None)
day1.columns=['id','time','danmu']
day1["time"]=pd.to_datetime(day1.time.values,unit="s",utc=True).tz_convert('Asia/Shanghai').strftime("%Y-%m-%d %H:%M")
# display(day1)
day1_shuaping=day1.drop_duplicates(subset=['id','time','danmu'])
print(len(day1),len(day1_shuaping))
display(day1_shuaping)
```

```python
time_group=day1_shuaping.groupby('time')
display(time_group.size().describe())

time_list=[]
danmu_list=[]
num_list=[]
for gn,gl in time_group:
    num=len(gl)
    time_list.append(gn)
    num_list.append(num)
    danmu_list.append(gl['danmu'].tolist())

dic={'time':time_list,'danmu':danmu_list,'num':num_list}
new_data=pd.DataFrame(dic)
# display(new_data)
# print(new_data['danmu'][0],type(new_data['danmu'][0]))

dele=[]
for index,row in new_data.iterrows():
# åˆ é™¤å¼¹å¹•è¿‡å¤šæˆ–è€…è¿‡å°‘çš„ å¹³å‡å€¼çš„300%
    if row['num']>1500 or row['num']<10:
        print(row)
        dele.append(index)

print(dele)
print(len(dele))
new_data.drop(index=dele,inplace=True)
new_data.to_csv(r'../data/final_room36252danmu0318.csv',index=None)
```
